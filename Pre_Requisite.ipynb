```python
# 2.Write Python code to multiply the above two matrices. Solve using list of lists and then use numpy. Compare the timing of both solutions. Which one is faster? Why?
import numpy as np
import time
arr_1 = [[1,2,3],[4,5,6]]
arr_2 = [[7,8],[9,10],[11,12]]

def multiplication_using_list(arr_1,arr_2):
    n = len(arr_1)
    k = len(arr_1[0])
    m = len(arr_2[0])
    
    C = [[0 for i in range(m)] for i in range(n)]

    for i in range(n):
        for j in range(m):
            for r in range(k):
                C[i][j] += arr_1[i][r] * arr_2[r][j]
    return C

start = time.time()
Compute = multiplication_using_list(arr_1,arr_2)
end = time.time()
print("Time of multiplication using list will be approximated : ",( end- start),"seconds")
# using numpy
array_1 = np.array(arr_1)
array_2 = np.array(arr_2)

start_1 = time.time()
Compute_1 = np.dot(array_1,array_2)
end_1 = time.time()
print("Time of multiplication using numpy will be approximated : ",( end_1- start_1),"seconds")

#conclusion : The time differences may appear very small for tiny matrices (like 2×3 × 3×2), but  NumPy is still faster.
```

```python
# 3. Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?
import random
import time
import statistics
import numpy as np

def naive_kth_largest(arr, k):
    temp = arr[:]
    for i in range(k-1):
        max_val = max(temp)
        temp.remove(max_val)
    return max(temp)

def naive_median(arr):
    n = len(arr)
    k = (n + 1) // 2 
    return naive_kth_largest(arr, k)

def quickselect(arr, k):
    if len(arr) == 1:
        return arr[0]
    pivot = random.choice(arr)
    lows = [el for el in arr if el < pivot]
    highs = [el for el in arr if el > pivot]
    pivots = [el for el in arr if el == pivot]

    if k <= len(lows):
        return quickselect(lows, k)
    elif k <= len(lows) + len(pivots):
        return pivots[0]
    else:
        return quickselect(highs, k - len(lows) - len(pivots))

def quickselect_median(arr):
    n = len(arr)
    k = (n + 1) // 2
    return quickselect(arr, k)

def benchmark():
    size = 5000
    data = random.sample(range(1, size*10), size)

    start = time.time()
    median1 = naive_median(data)
    t1 = time.time() - start
    print(f"Naive median: {median1}, Time: {t1:.4f} seconds")

    start = time.time()
    median2 = quickselect_median(data)
    t2 = time.time() - start
    print(f"Quickselect median: {median2}, Time: {t2:.4f} seconds")

    start = time.time()
    median3 = statistics.median(data)
    t3 = time.time() - start
    print(f"Statistics.median: {median3}, Time: {t3:.4f} seconds")

    start = time.time()
    median4 = np.median(data)
    t4 = time.time() - start
    print(f"Numpy median: {median4}, Time: {t4:.4f} seconds")

    print(f"All agree? {median1 == median2 == median3 == int(median4)}")

benchmark()


# -The Quickselect algorithm can find the k-th smallest element in average O(n) time.
# -This is much better than O(n^2).

```

```python
#5.Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y
import numpy as np
from scipy.optimize import approx_fprime
import jax
import jax.numpy as jnp

def f_numpy(vec):
    x, y = vec
    return (x**2) * y + (y**3) * np.sin(x)

def f_jax(vec):
    x, y = vec
    return (x**2) * y + (y**3) * jnp.sin(x)

def analytical_gradient(vec):
    x, y = vec
    df_dx = 2*x*y + y**3 * np.cos(x)
    df_dy = x**2 + 3*(y**2) * np.sin(x)
    return np.array([df_dx, df_dy])


grad_f_jax = jax.grad(f_jax, argnums=0)  # returns gradient w.r.t. input vector

epsilon = np.sqrt(np.finfo(float).eps)

np.random.seed(42)
for i in range(5):
    point = np.random.uniform(low=-5, high=5, size=2)

    grad_scipy = approx_fprime(point, f_numpy, epsilon)
    grad_jax = np.array(grad_f_jax(point))
    grad_analytical = analytical_gradient(point)

    print(f"\nTest Point {i+1}: x = {point[0]:.4f}, y = {point[1]:.4f}")
    print(f"Analytical Gradient:  {grad_analytical}")
    print(f"SciPy Approx Gradient: {grad_scipy}")
    print(f"JAX Auto-Gradient:     {grad_jax}")

    assert np.allclose(grad_analytical, grad_jax, atol=1e-6)
    assert np.allclose(grad_analytical, grad_scipy, atol=1e-6)
    print(" Gradients match.")
```

```python
#6. use sympy to confirm that you obtain the same gradient analytically.
import sympy as sp
x, y = sp.symbols('x y')
f = (x**2)* y + (y**3)* sp.sin(x)

df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("∂f/∂x =", df_dx)
print("∂f/∂y =", df_dy)

expected_df_dx = 2 * x * y + (y**3) * sp.cos(x)
expected_df_dy = (x**2)+ 3 * (y**2) * sp.sin(x)

assert sp.simplify(df_dx - expected_df_dx) == 0
assert sp.simplify(df_dy - expected_df_dy) == 0
print("Symbolic gradient matches the manual analytical gradient.")
```

```python
#8. Store the same information using Python classes. We have an overall database which is a list of year objects. Each year contains a list of branches. Each branch contains a list of students. Each student has some properties like name, roll number and has marks in some subjects.
class Student:
    def __init__(self, name, roll_number, marks):
        """
        marks: dict where keys are subject names and values are marks.
        Example: {'Math': 90, 'Physics': 85}
        """
        self.name = name
        self.roll_number = roll_number
        self.marks = marks

    def __repr__(self):
        return f"Student(name={self.name}, roll_number={self.roll_number}, marks={self.marks})"


class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []  

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch(name={self.name}, students={self.students})"


class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []  

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"Year(year={self.year}, branches={self.branches})"


class Database:
    def __init__(self):
        self.years = []  # List of Year objects

    def add_year(self, year):
        self.years.append(year)

    def __repr__(self):
        return f"Database(years={self.years})"
```

```python
#13.Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

import numpy as np

def my_argmin(arr):
    arr = np.asarray(arr)    
    min_index = 0
    min_value = arr[0]
    
    for i in range(1, len(arr)):
        if arr[i] < min_value:
            min_value = arr[i]
            min_index = i
    
    return min_index

# Example
arr = np.array([5, 2, 8, 1, 7])
print("My argmin:", my_argmin(arr))
print("NumPy argmin:", np.argmin(arr))
```
